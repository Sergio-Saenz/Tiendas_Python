### Importación de datos

#1. Análisis de facturación



```python
# Ventas de la 1 Tienda
precio = round(tienda["Precio"].sum())
print("Cantidad total de unidades vendidas en la tienda_1:", f"{precio:,.2f}")

Costo_Envio = round(tienda["Costo de envío"].sum())
print("Cantidad total de unidades vendidas en la tienda_1:", f"{Costo_Envio:,.2f}")

Total_Facturado= round(precio + Costo_Envio)
print("Total facturado en la tienda_1:",f"{Total_Facturado:,.2f}")
```

# 2. Ventas por categoría

```python
# Categorias de la 1 Tienda
from matplotlib import pyplot as plt
Contador_1 = {}

for i in range(len(tienda["Categoría del Producto"])):
    categoria_1 = tienda["Categoría del Producto"][i]
    if categoria_1 in Contador_1:
        Contador_1[categoria_1] += 1
    else:
        Contador_1[categoria_1] = 1

# Ordenamos las categorías de mayor a menor cantidad
categorias_ordenadas = sorted(Contador_1.items(), key=lambda x: x[1], reverse=True)

categorias = [item[0] for item in categorias_ordenadas]
cantidades = [item[1] for item in categorias_ordenadas]

# Mostramos el resultado
print("Categorías más repetidas:")
for categoria, cantidad in categorias_ordenadas:
    print(f"{categoria}: {cantidad}")

plt.bar(x=categorias,height=cantidades)
plt.xticks(rotation=90)
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Productos mas y menos vendidos de la tienda 1
Contador = {}

for i in range(len(tienda["Producto"])):
    productos = tienda["Producto"][i]
    if productos in Contador:
        Contador[productos] += 1
    else:
        Contador[productos] = 1

# Ordenamos las calificaciones de mayor a menor cantidad
cantidad_productos = sorted(Contador.items(), key=lambda x: x[1], reverse=True)

# Mostramos el resultado
print("Producto | Cantidad")
for productos, cantidades in cantidad_productos:
    print(f"{productos} ---> {cantidades}")

producto = [item[0] for item in cantidad_productos]
cantidad = [item[1] for item in cantidad_productos]

plt.bar(x=producto,height=cantidad)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
